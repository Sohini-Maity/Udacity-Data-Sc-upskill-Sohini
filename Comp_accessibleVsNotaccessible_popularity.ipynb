```python
import os
import pandas as pd 
import matplotlib.pyplot as plt

# file path #
listings = r'C:\Users\XXX\Desktop_files\Udacity data sc files\listings.csv'

#read file#
df_listing=pd.read_csv(listings)

# reading Super host Properties #
df_listing=df_listing[df_listing['host_is_superhost']=='t'][['id', 'name','amenities','property_type','room_type','reviews_per_month','price','security_deposit','cleaning_fee','availability_30','cancellation_policy']]
  
# checking the properties wheelchair accessibility#
df_listing['accessibilty_check']=df_listing['amenities'].str.contains('"Wheelchair Accessible"')

# checking the properties got more reviews in a month #
df_check = df_listing.fillna(0).sort_values(['reviews_per_month'], ascending=[False])

## Plottng the data to comapre accessibile/non access property functioning  w.r.t use in 1 month#
df_check.plot.hist(column=["availability_30"], by="accessibilty_check")

```
