```python
import os
import pandas as pd 
import matplotlib.pyplot as plt

## reading data from calendar file#
calendar=r'C:\Users\Sohini.Maity\OneDrive - Shell\Desptop_files\Udacity data sc files\calendar.csv'

df_calendar=pd.read_csv(calendar)

# datatype change and deriving Year and Month Column##
df_calendar['date_time']=pd.to_datetime(df_calendar['date'])
df_calendar['Month'] = df_calendar['date_time'].dt.month
df_calendar['Year'] = df_calendar['date_time'].dt.year

# month on which most property not available ##
df_calendar=df_calendar[(df_calendar['available']=='f') & (df_calendar['Year']==2016)][['Month','Year']]
df=df_calendar.value_counts(['Month','Year']).reset_index(name='count').sort_values(['count'], ascending=[False])

## plotting to get insight what time of the year properties are popular##
df.plot(y='count',x='Month', kind='bar', legend=False)

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Popular Time of the Year')

 
```
